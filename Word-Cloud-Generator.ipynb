```python
from textblob import TextBlob
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from collections import Counter
from IPython.display import display
import ipywidgets as widgets
```

```python
nltk.download('punkt')
```

```python
text_area = widgets.Textarea(
    value='Insira o texto aqui',
    placeholder='Digite algo',
    description='Texto:',
    disabled=False,
    layout=widgets.Layout(width='auto', height='200px')
)

display(text_area)

def process_text(text):
    blob = TextBlob(text)
    palavras = blob.words

    stop_words = [
        "o", "a", "e", "que", "de", "para", "em", "dos", "das", "com", "no", "na", "se", "me", "te",
        "se", "lhe", "nos", "vos", "lhes", "um", "uns", "uma", "umas", "do", "da", "dos", "das",
        "com", "no", "na", "se", "me", "te", "se", "lhe", "nos", "vos", "lhes", "como", "é", "por", "os", "todo", "os", "as", "todos", "ele", "era", "ao", "à", "4", "294", "5", "já", "não", "entre", "logo", "mão", "só", "são", "foi", "outros", "está", "site", "r", "aos", "disse", "eles", "até", "seus", "sua", "dia", "sem", "mas", "sobre", "também",
        "deu", "mais", "Porque", "porque", "pela", "faz", "fez", "fazia", "teria", "isso", "vez"
    ]

    palavras_sem_stopwords = [palavra.lower() for palavra in palavras if palavra.lower() not in stop_words]

    return palavras_sem_stopwords

def on_button_click(b):
    text = text_area.value
    palavras_sem_stopwords = process_text(text)

    # Contar a frequência das palavras
    contador = Counter(palavras_sem_stopwords)

    # Exibir o gráfico de barras
    labels, values = zip(*contador.items())
    plt.figure(figsize=(10, 5))
    plt.bar(labels, values)
    plt.xticks(rotation='vertical')
    plt.show()

    # Criar e exibir a nuvem de palavras
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(contador)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()

# Botão para processar o texto
button = widgets.Button(description="Processar Texto")
button.on_click(on_button_click)
display(button)
```
